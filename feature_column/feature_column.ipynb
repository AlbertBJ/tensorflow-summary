首先大家一定 要记得 神经网络 能够处理什么数据类型，答案是 数字类型，所以 无论您的 原始数据集 是什么格式，最终 都必须 要转换成 数字类型 才能送入 最终的model.tf官网提供了 feature_column包，就是 为了 将 其它 类型 转换为 数字类型的。下面 分别说明

namespace： tf.feature_column

feature_column是原始数据和estimator模型之间的桥梁，如下图所示：



tf主要有两类column,一类是生成dense tensor的Dense column . 另一类是 生成sparse tensor 的Categorical column,如下所示：








# input layer


```python
tf.feature_column.input_layer(
    features,
    feature_columns,
    weight_collections=None,
    trainable=True,
    cols_to_vars=None,
    cols_to_output_tensors=None
)
使用 input_layer 作为model的一个 input layer
features:字典，最主要的是 dict的key一定要与 feature_columns的key一致，后续才能 才能根据key进行匹配
feature_columns：改参数必须是 继承于DenseColumn的numeric_column, embedding_column, bucketized_column, indicator_column。如果feature是类别的，那么必须先
                 用embedding_column or indicator_column封装一下使用
```

```python
# 1. numeric_column

tf.feature_column.numeric_column(
    key,
    shape=(1,),
    default_value=None,
    dtype=tf.dtypes.float32,
    normalizer_fn=None
)

参数说明：
key:唯一确定输入特征的字符串，也就是说 该字段 用来标识 这个 特征
shape：该特征的 形状，返回的 tensor shape :batch+shape:比如 shape=1, 则返回 shape是 batch*1 
default_value 默认值
dtype：数据类型
normalizer_fn：做一些 normalization操作，比如 标准化。normalizer 函数 input:tensor；output:tensor
```

```python
x_train.keys()
```

```python
feature_columns=[tf.feature_column.numeric_column(key,default_value=0) for key in x_train.keys()]
```

```python
feature_columns
```

```python
# 使用 input_layer 作为model的一个 input layer
inn=tf.feature_column.input_layer(features,feature_columns)
print(inn)
```

```python
with tf.Session() as sess:
    print(sess.run(inn))
```

# 使用normalizer_fn

```python

# 比如标准化

mean=x_train['Sepallength'].mean()
std=x_train['Sepallength'].std()
print('mean is {},std is {}'.format(mean,std))
def normalization(mean,std):
    def normalizer_fn(x):
        return (x-mean)/std   
    return normalizer_fn
```

```python
feature=tf.feature_column.numeric_column('Sepallength',default_value=0,normalizer_fn=normalization(mean,std))  
feature
```

```python
in_Sepallength=tf.feature_column.input_layer(features,[feature])
in_Sepallength
```

```python
with tf.Session() as sess:
    print(sess.run(in_Sepallength))
```

## tf.feature_column.categorical_column_with_identity

可以理解为 将类型为int的特征，转换为 one-hot类型的 dense tensor

```python
tf.feature_column.categorical_column_with_identity(
    key,
    num_buckets,
    default_value=None
)
首先 它返回的是 唯一值的 类别column(categoricalColumn)

key：参数同上
num_buckets:限制输入输出的数据范围。范围是[0,num_buckets)，还有就是 使用 这个函数，对应的feature value必须是 int类型
default_value:默认值，该值必须在 [0,num_buckets)范围内。如果不设置 默认值，则 会抛出异常（如果 数据不在  [0,num_buckets)范围内）                            
```

```python
columns_name=['id','cat']
data=pd.read_csv('data/categorical_column.csv',names =columns_name)
```

```python
features_new = {key:np.array(value) for key,value in dict(data).items()}
features_new
```

```python
ci=tf.feature_column.categorical_column_with_identity(   'id',   10)
ci_int=tf.feature_column.indicator_column(ci) # 为什么 要添加 这个步骤呢？因为 input_layer的入参是有要求的，
                                              # 如上所说必须是继承于DenseColumn的类型,详看 下面的 解释
```

```python
in_ci=tf.feature_column.input_layer(features_new,[ci_int])
in_ci
```

```python
with tf.Session() as sess:
    print(sess.run(in_ci))
```

## tf.feature_column.indicator_column

tf.feature_column.indicator_column(categorical_column)

参数：
必须是categorical_column，如上面图可知 ，只有 categorical_column_with_* ,crossed_column 以及bucketized_column 类型的column 才可以使用该函数

作用：将 categorical_column表示成 multi-hot形式的 dense tensor。

下面 官方 还特意 提到了 一点就是：
  indicator_column 可以将 任意的 categorical_column_with_*作为参数（同我上面说的一致），
  但是，如果 说  buckets/unique（分箱，或者 ont-hot） 的值 比较大的时候，还是 建议使用 embedding_column




## tf.feature_column.categorical_column_with_vocabulary_file


tf.feature_column.categorical_column_with_vocabulary_file(
    key,<br>
    vocabulary_file,<br>
    vocabulary_size=None,<br>
    num_oov_buckets=0,<br>
    default_value=None,<br>
    dtype=tf.dtypes.string<br>
)

作用：将 表示类别 的 feature转换为CategoricalColumn tensor

参数：
key:同上

vocabulary_file; 字典文件

vocabulary_size：字典文件中 元素个数，该值不能大于 字典元素数；如果该值 小于 字典元素数，则字典中后续的 元素将被忽略。默认情况下，该值 会                  被 自动设置为 字典元素数

num_oov_buckets： 就是说 如果 value值 超出了 字典 ，后续 应该用 什么 值 来填充，基于 hash算法，填充值在                         [vocabulary_size,vocabulary_size+num_oov_buckets)范围内。 注意 该值与 default_value不共存(不能同时 存在，否则，会报错)

default_value：默认-1，就是说 当 feature value的值 不在 字典中时，也就是oov,使用 一个 整数 默认值 来填充


dtype： feature type ,目前仅支持 string 和 integer



```python
cate_vf=tf.feature_column.categorical_column_with_vocabulary_file('cat',
                                                                  vocabulary_file='data/vocabulary_file.csv',
                                                                  vocabulary_size=2,
                                                                  num_oov_buckets=3,
                                                                  
                                    
                                                                 )

ind_cat=tf.feature_column.indicator_column(cate_vf)
ind_cat

inp_cat=tf.feature_column.input_layer(features_new,[ind_cat])
inp_cat

with tf.Session() as sess:
    # 在此处 必须使用 tf.tables_initializer来初始化 lookuptable
    sess.run(tf.tables_initializer())
    print(sess.run(inp_cat))
```

## tf.feature_column.categorical_column_with_hash_bucket

tf.feature_column.categorical_column_with_hash_bucket(<br>
    key,<br>
    hash_bucket_size,<br>
    dtype=tf.dtypes.string<br>
)<br>
作用：利用hash算法，将 特征值类型为 integer/string的 特征 进行 分箱操作<br>
参数：<br>
key:同上<br>
hash_bucket_size: int类型 且>1 <br>
dtype: 特征的类型，目前仅支持 string 和integer


```python
cate_hb=tf.feature_column.categorical_column_with_hash_bucket('id',  hash_bucket_size=5,dtype=tf.int64)
ind_cat_hb=tf.feature_column.indicator_column(cate_hb)
inp_cat_hb=tf.feature_column.input_layer(features_new,[ind_cat_hb])
with tf.Session() as sess:
    print(sess.run(inp_cat_hb))
```

## tf.feature_column.bucketized_column


tf.feature_column.bucketized_column(<br>
    source_column,<br>
    boundaries<br>
)
<br>分箱，目的是 离散化 数据，比如 将 年龄 字段 划分为 少年，青年，壮年以及老年<br>
参数：<br>

source_column: numeric_column 
<br>
boundaries: 边界列表，形成 左闭右开 区间，比如 [1,2,3],那么拆分后的边界将是 (-inf,1),[1,2),[2,3),[3,+inf) 四个区间<br>


```python
bc_fc=tf.feature_column.numeric_column('id',default_value=0)
bucket=tf.feature_column.bucketized_column(bc_fc,boundaries=[3,5,8])
print(bucket)
input_bucket=tf.feature_column.input_layer(features_new,[bucket])
with tf.Session() as sess:
    print("**********")
    print(input_bucket)
    print('*********')
    print(sess.run(input_bucket))
```

## tf.feature_column.crossed_column
<br>
tf.feature_column.crossed_column(<br>
    keys,<br>
    hash_bucket_size,<br>
    hash_key=None<br>
)<br>
说明：该函数的作用主要是 通过笛卡尔积 产生 新的特征；然后经过hash算法 产生最终特征：Hash(特征的笛卡尔积) % hash_bucket_size<br>
参数：<br>
keys：指明 那些 feature 需要进行 合并 形成新的特征，注意，该值的类型可以是 string(也就是数据集的column),也可以是 经过变换后的 categoricalColumn, but,不支持 hashed categorical<br>
hash_bucket_size: 分箱的 长度，可以理解为 合并后 数据的列数<br>




```python
id_x_cat = tf.feature_column.crossed_column(['id', 'cat'], 5)
fc_ic=tf.feature_column.indicator_column(id_x_cat)
inp_cc=tf.feature_column.input_layer(features_new,[fc_ic])
inp_cc
```

```python
with tf.Session() as sess:
    print(sess.run(inp_cc))
```


## tf.feature_column.embedding_column<br>

tf.feature_column.embedding_column(<br>
    categorical_column,<br>
    dimension,<br>
    combiner='mean',<br>
    initializer=None,<br>
    ckpt_to_load_from=None,<br>
    tensor_name_in_ckpt=None,<br>
    max_norm=None,<br>
    trainable=True<br>
)<br>

该方法和indicator_column一样，只接受 categorical_column,目的是将 稀疏矩阵转换为稠密矩阵， word2vec<br>
参数：<br>
categorical_column:入参categorical_column_with_*的返回<br>
dimension: embedding 的维度，一般计算规则是 类别开4次方，但是也可以根据需要自行设置<br>
combiner: 多个 vector的 组合方式，有 mean(default),sqrtn以及sum<br>
initializer:embedding matrix的初始化值，默认 均值0，标准差 1/sqrt(dimension)的tf.truncated_normal_initializer <br>
ckpt_to_load_from以及tensor_name_in_ckpt 主要是为了使用pre-trained embedding matrix <br>
max_norm: if not 'None'，则使用l2 归一化<br>
trainable: 是否可训练的
